```python
import torch
import torch.nn as nn
import torchvision.models as models

# -------------------------
# 1. Device setup
# -------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# -------------------------
# 2. Recreate your model architecture (3 channels)
# -------------------------
model = models.resnet18(weights=None)  # no pretrained weights
num_classes = 10  # EuroSAT has 10 land use classes
model.fc = nn.Linear(model.fc.in_features, num_classes)

# -------------------------
# 3. Load weights
# -------------------------
model_path = r"C:\Users\Ekaansh\OneDrive\Desktop\AB\vs code\JS\projects\satelite\test 2\eurosat_resnet18.pth"
state_dict = torch.load(model_path, map_location=device)
model.load_state_dict(state_dict)

# -------------------------
# 4. Move to device and eval
# -------------------------
model = model.to(device)
model.eval()
```
